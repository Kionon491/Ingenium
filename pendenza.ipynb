## Pendenza di ua curva

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definizione della funzione $f(x)$

```python
def f(x):
    return(x**3*x**2-5*x+2)
```

```python
f(0)
```

Genero il vettore degli argomenti $x$

```python
x=np.linspace(-10,10,1004)
print(x)
```

Genero il vettore delle immagini $y$

```python
y=f(x)
print(y)
```

```python
plt.plot(x, y)
plt.xlim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico della funzione y=f(x)')
plt.grid(True)
plt.ylim(-10,10)
plt.show()
```

Funzione endenza $m(x)$

```python
def m(x):
    dx=1e-5
    return (f(x+dx)-f(x-dx))/(2*dx)
```

```python
plt.plot(x, y, label='f(x)')
plt.plot(x, m(x), label='m(x) (pendenza)', color='red')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico di f(x) e m(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
coeff_m = np.polyfit(x, m(x), deg=2)
coeff_m_rounded = np.round(coeff_m, 4)
print("Coefficenti della curva m(x) di secondo grado arrotondati a 4 cifre significative:", coeff_m_rounded)
```
